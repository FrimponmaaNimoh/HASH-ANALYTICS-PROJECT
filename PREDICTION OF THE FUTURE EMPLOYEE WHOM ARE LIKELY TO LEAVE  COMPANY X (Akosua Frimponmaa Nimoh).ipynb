#### FINDING EXISTING EMPLOYEES LIKELY TO LEAVE COMPANY X

##### Determining Factors

1) An employee within the sales,support,technical department.

2) An employee with low salary.

3) An employee with an average job satisfaction of 0.44.

4) An employee with no promotions after working 3-5 years.

5) An employee with 0 work accidents.


```python

condition_1_bool= ((existing_employees["dept"]=="sales")|(existing_employees["dept"]=="support")|(existing_employees["dept"]=="technical")) & (existing_employees["salary"]=="low" ) & (existing_employees["promotion_last_5years"]==0)&(existing_employees["satisfaction_level"]<=0.44)&(existing_employees["Work_accident"]==0)&(existing_employees["time_spend_company"]==3)
likely_to_leave_employee = existing_employees.loc[condition_1_bool].sort_values("satisfaction_level").reset_index(drop=True)
likely_to_leave_employee
```

#### MOST PRONE TO LEAVE

An employee surely within the sales department, with low salary , an average job satisfaction of 0.44 and has no promotions after working 3 years



```python
most_prone_condition_bool= (likely_to_leave_employee["dept"]=="sales")
most_prone_to_leave_employee = likely_to_leave_employee.loc[most_prone_condition_bool].sort_values(["satisfaction_level","number_project"]).reset_index(drop=True)
most_prone_to_leave_employee
```
